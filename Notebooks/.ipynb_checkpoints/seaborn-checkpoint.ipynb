```python
#Exercise create a sub dataframe of the following variables and see the correlation

# LotArea
# YearBuilt
# PoolArea
# SalePrice
# GarageArea


df1 = df[['LotArea', 'YearBuilt','SalePrice', 'GarageArea']]
df1.head

df1.shape
```

```python
# Compute the correlation matrix
corr = df1.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=np.bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
# Let us first normalize one column only 


df1["SalePrice"]=((df1["SalePrice"]-df1["SalePrice"].min())/(df1["SalePrice"].max()-df1["SalePrice"].min()))

df1["SalePrice"].describe()

```

```python
# Let us normalize all the columns in the df

df1=((df1-df1.min())/(df1.max()-df1.min()))

df1.head()
```

```python
# Let us view the distribution of these variables 

sns.distplot(df1['SalePrice'], hist=False, rug=True);
```

```python
# What if we want to plot all togethetr 

for col in df1.columns:
    sns.distplot(df1[col], hist=False, rug=True, label=col)

# This is a good way to plot many graphs on the same chart
```

```python
for col in ["YearBuilt", "SalePrice"]:
    sns.distplot(df1[col], hist=False, rug=True, label=col)
    
```

```python
# Let us plot a joint distribution 

g = sns.jointplot(x="SalePrice", y="YearBuilt", data=df1, kind="kde", color="m")
g.plot_joint(plt.scatter, c="w", s=30, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$SalePrice$", "$YearBuild$");
```

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
df1.head()

```

```python
# We need to reorder the columns so Sales price is the last column
# SalesPrice will be our Y and everything else will be X. 
# We will build a neural network for Y=f(X) to predict the 
# SalePrice

cols = ['LotArea', 'YearBuilt', 'GarageArea','SalePrice']
df1 = df1[cols]

df1
```

```python
# Split the data into X and Y
# But before we do this we will need to 

# convert the array into a numpy array
arr = df1.to_numpy()

# separate X and Y
X = arr[:,0:3]
Y = arr[:,3]
```

```python
X
```

```python
Y
```

```python
# define the keras model
model = Sequential()
model.add(Dense(12, input_dim=3, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1))
```

```python
# compile the keras model
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mean_absolute_error', 'mean_squared_error'])
```

```python
history = model.fit(X, Y, validation_split=0.25, epochs=100, batch_size=32, verbose=1)
```

```python
# Plot training & validation accuracy values
plt.plot(history.history['mean_absolute_error'])
plt.plot(history.history['val_mean_absolute_error'])
plt.plot(history.history['mean_squared_error'])
plt.plot(history.history['val_mean_squared_error'])
plt.title('Mean Absolute Error')
plt.ylabel('Mean Absolute Error')
plt.xlabel('Epoch')
plt.legend(['MAE Train', 'MAE Validation', 'MSE Train', 'MSE Validation'], loc='upper left')

plt.show()
```

```python
print('\n# Generate predictions for 3 samples')
predictions = model.predict(X)
print('predictions shape:', predictions.shape)
print(type(predictions))

```

MAE (abs (actual-predicted)) and MSE (actual-predicted)^2 measure slightly different types of errors. Because MSE squares the difference, it will enhance large errors. So use MSE if you care about getting rid of large errors. 

```python
# Now let us visuall see how far actual values are from the predicted ones
# Need to flaten because it is an array within an array
test_predictions = model.predict(X).flatten()

sns.set_style('ticks') 

a = plt.axes(aspect='equal')
plt.scatter(Y, test_predictions)
plt.xlabel('True Values')
plt.ylabel('Predictions')
lims = [0, 1]
plt.xlim(lims)
plt.ylim(lims)
_ = plt.plot(lims, lims)
```

```python
# Let us fit a regression line to see that actual error
# Make a nice data frame from the two rows of data

def time_built(x):
    if (x < 0.5): return "early"
    else: return "late"
    
    
yb = df1['YearBuilt'].apply(time_built).to_numpy()  # need to have a numpy array for the next step.

d = {'actual':Y, 'predicted':test_predictions, 'year_built':yb} 
d
```

```python
# Create a pandas dataframe
rframe = pd.DataFrame(d)
sns.set_style('whitegrid') 
sns.lmplot(x ='actual', y ='predicted', data = rframe, hue="year_built", line_kws={"color": "red"})
```
