```python
## Section I: Data Loading and Manipulation ##
```

```python

# TODO: load the given input CSV file 'Telco-Customer-Churn.csv' into a DataFrame using pandas
# Ensure that the file 'Telco-Customer-Churn.csv' is saved in the same directory as this notebook

# HINT: You may use pandas 'read_csv' function

# ENTER YOUR CODE BELOW:-

df = pd.read_csv('Telco-Customer-Churn.csv')



```

```python

# TODO: Show the top 5 rows from your DataFrame

# HINT: You may use the DataFrame 'head' function

# ENTER YOUR CODE BELOW:-

df.head()



```

```python

# TODO: Print the summary of your DataFrame to list all the different column names and their types

# HINT: You may use the DataFrame 'info' function

# ENTER YOUR CODE BELOW:-


df.info()

```

```python

# TODO: As it can be seen in the previous result, the 'TotalCharges' column is of type 'object', but
#       since this column actually contains numeric values, you are required to:
#       1) Convert this column to numeric type,
#       2) and then, re-print the updated DataFrame columns and types summary

# HINT: You may use pandas 'to_numeric' function and then the DataFrame 'info' function

# ENTER YOUR CODE BELOW:-

#use errors = coerce to deal with bad values so it can be converted to NaN
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors = 'coerce')
df.info()

```

```python

# TODO: As you can notice in the previous result, in the 'TotalCharges' column there are 7032 rows compared to the 7043
#       rows for others, which indicates that there may be "11" missing or blank values. In data preparation, there are
#       different ways of handling the missing values, for example we may replace them with statistical values such as
#       median, or set them to 0, or we may even delete them if the missing values count is relatively smaller compared
#       to the overall size of our dataset.
#       For this exercise, you are required to:
#       1) Delete those 11 missing values rows (we are deleting as they are only a small number of rows as explained above).
#       2) Re-print the DataFrame columns and types summary to show the updated number of rows

# HINT: You may use the DataFrame 'dropna' function and then the DataFrame 'info' function

# ENTER YOUR CODE BELOW:-


df = df.dropna()
df.info()
```

```python

# TODO: Delete also the 'customerID' column, since we would not use this for our further analysis as it is only
#       a unique identifier. After deleting, print the top 5 rows from your DataFrame again (this will now exclude
#       the 'customerID' column)

# HINT: You may use the DataFrame 'drop' function for deleting and then the 'head' function to display the top 5 rows

# ENTER YOUR CODE BELOW:-

df = df.drop(columns=['customerID'])

df.head()
```

```python

## Section II: Data Exploration and Visualization ##
```

```python

# TODO: Plot the distribution of the target column 'Churn' values in your dataset as vertical bar chart.

# HINT: You may use seaborn 'countplot' function, and plot 'Churn' values on the horizontal axis.

# ENTER YOUR CODE BELOW:-

chart = sns.countplot(x="Churn", data=df)


chart.set_xticklabels(chart.get_xticklabels(), rotation=45)

# Show plot
plt.show()




```

```python

# TODO: To understand the 'Churn' column trend, review the pair-wise relationship between the numerical columns
#       'tenure', 'MonthlyCharges' and 'TotalCharges' against the 'Churn' column.

# HINT: You may use seaborn 'pairplot' function, with variables ('vars=') as columns 'tenure', 'MonthlyCharges',
#       'TotalCharges', and 'hue' as the 'Churn' column.

# ENTER YOUR CODE BELOW:-

chart = sns.pairplot(vars = ['tenure', 'MonthlyCharges', 'TotalCharges'], hue = 'Churn', data=df)
chart

```

```python

# TODO: Prepare the dataset for further analysis by converting selected text-based columns into integer types.
#       Convert the 'gender' column values into integers by replacing the values 'Male' with '1' and 'Female' with '0'.
#       After converting, print the top 5 rows from your DataFrame again to check the updated 'gender' values

# HINT: You may use the DataFrame 'gender' attribute and use 'if else' statement to replace '1' if value is 'Male'
#       and '0' if otherwise. Then use the DataFrame 'head' function to print the top 5 rows.

# ENTER YOUR CODE BELOW:-

df['gender'] = df['gender'].map(dict(zip(['Male','Female'],[1,0])))

df.head()
                                
                                




```

```python

# TODO: Convert the remaining columns 'Churn', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'OnlineSecurity'
#       'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling' to integers.

# HINT: You may use the DataFrame attributes each of the above columns to replace with '1' if value is 'Yes' and '0'
#       if value is 'No' or '-1' if otherwise. 

# ENTER YOUR CODE BELOW:



df['Churn'] = df['Churn'].apply(lambda x: 1 if x == 'Yes' else (0 if x == "No" else -1))
df['Partner'] = df['Partner'].apply(lambda x: 1 if x == 'Yes' else (0 if x == "No" else -1))
df['Dependents'] = df['Dependents'].apply(lambda x: 1 if x == 'Yes' else (0 if x == "No" else -1))
df['PhoneService'] = df['PhoneService'].apply(lambda x: 1 if x == 'Yes' else (0 if x == "No" else -1))
df['MultipleLines'] = df['MultipleLines'].apply(lambda x: 1 if x == 'Yes' else (0 if x == "No" else -1))
df['OnlineSecurity'] = df['OnlineSecurity'].apply(lambda x: 1 if x == 'Yes' else (0 if x == "No" else -1))
df['OnlineBackup'] = df['OnlineBackup'].apply(lambda x: 1 if x == 'Yes' else (0 if x == "No" else -1))
df['DeviceProtection'] = df['DeviceProtection'].apply(lambda x: 1 if x == 'Yes' else (0 if x == "No" else -1))
df['TechSupport'] = df['TechSupport'].apply(lambda x: 1 if x == 'Yes' else (0 if x == "No" else -1))
df['StreamingTV'] = df['StreamingTV'].apply(lambda x: 1 if x == 'Yes' else (0 if x == "No" else -1))
df['StreamingMovies'] = df['StreamingMovies'].apply(lambda x: 1 if x == 'Yes' else (0 if x == "No" else -1))
df['PaperlessBillin'] = df['PaperlessBilling'].apply(lambda x: 1 if x == 'Yes' else (0 if x == "No" else -1))



df.head()







# df = df.replace([1, 0, -1], ['Yes', 'No', ''])
# df
```

```python

# TODO: Review the 'Churn' probabilities for customers with different types of 'Contract' by plotting a vertical bar chart
#       with 'Contract' type on the horizontal axis and 'Churn Probability' on the vertical axis.

# HINT: You may use seaborn 'catplot' function.

# ENTER YOUR CODE BELOW:-

churn_probs = pd.DataFrame(df.groupby('Contract')['Churn'].size().div(len(df)))


print(churn_probs.shape)
print(churn_probs)

#sns.countplot(x="Contract", data=df)



chart = sns.barplot(x = "Contract", y = df.groupby('Contract')['Churn'].size().div(len(df)), data = df)

chart.set_xticklabels(chart.get_xticklabels(), rotation=45)

# Show plot
plt.show()

```
